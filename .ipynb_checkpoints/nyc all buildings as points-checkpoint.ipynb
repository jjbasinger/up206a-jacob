# Notebook for getting all of the NYC buildings per Census Tract

This notebook is intended for getting the total number of buildings per census tract in NYC.

```python
import pandas as pd
import geopandas as gpd
import sodapy
from sodapy import Socrata
import matplotlib.pyplot as plt
```

API for NYC open data:

```python
client = Socrata("data.cityofnewyork.us", '9llM0ejMVTKfRxS1XlvL7gXjU') 
#included my app token to reduce throttle limits
```

I utilized interesting dataset that has all of the building footprints in NYC: https://data.cityofnewyork.us/Housing-Development/Building-Footprints/nqwf-w8eh.  Additional documentation can be found here: https://github.com/CityOfNewYork/nyc-geo-metadata/blob/master/Metadata/Metadata_BuildingFootprints.md.  I used the API to call the portion of the dataset that only has point representations of each building, instead of a shapefile for each footprint.

Accessing the point data.

```python
buildingresults = client.get("3577-dfxj", content_type='geojson',
                             select = 'the_geom',
                             limit = 1100000)
```

```python
buildings = gpd.GeoDataFrame.from_features(buildingresults)
```

```python
buildings = buildings.set_crs(4326)
```

```python
buildings.crs
```

```python
buildings.info()
```

Now we have a dataset of points for all of the buildings in NYC.

# Joining the building data to census tracts

Recreating code already used from another notebook in which we first worked with census tract data.  Importing, renaming, and cleaning the data.

```python
censustracts = gpd.read_file('../up206a_groupproject/currentdata/nyctracts.json')
```

```python
censustracts = censustracts[['BoroName', 'CT2020', 'Shape__Area', 'Shape__Length', 'geometry']]
```

```python
censustracts.columns = ['borough', 'census_tract', 'Shape__Area', 'Shape__Length', 'geometry']
```

```python
censustracts['census_tract'] = censustracts['census_tract'].astype('float')
```

```python
censustracts.plot()
```

```python
censustracts.info()
```

Unfortunately, the census tract numbers are repeated in each borough since they are independent counties.  Because of this, I will break the census tract data into individual datasets and work using those.

```python
manhattantracts = censustracts[censustracts['borough']=='Manhattan']
brooklyntracts = censustracts[censustracts['borough']=='Brooklyn']
bronxtracts = censustracts[censustracts['borough']=='Bronx']
statenislandtracts = censustracts[censustracts['borough']=='Staten Island']
queenstracts = censustracts[censustracts['borough']=='Queens']
```

Spatial joining the building point data with each of the borough census tracts.

```python
mjoingdf = gpd.sjoin(buildings, manhattantracts)
brjoingdf = gpd.sjoin(buildings, brooklyntracts)
bxjoingdf = gpd.sjoin(buildings, bronxtracts)
sijoingdf = gpd.sjoin(buildings, statenislandtracts)
qjoingdf = gpd.sjoin(buildings, queenstracts)
```

```python
qjoingdf.info()
```

Getting the quantity of buildings per tract in each borough.

```python
buildings_by_mtracts = mjoingdf.census_tract.value_counts().rename_axis('census_tract').reset_index(name='buildings_count')
buildings_by_brtracts = brjoingdf.census_tract.value_counts().rename_axis('census_tract').reset_index(name='buildings_count')
buildings_by_bxtracts = bxjoingdf.census_tract.value_counts().rename_axis('census_tract').reset_index(name='buildings_count')
buildings_by_sitracts = sijoingdf.census_tract.value_counts().rename_axis('census_tract').reset_index(name='buildings_count')
buildings_by_qtracts = qjoingdf.census_tract.value_counts().rename_axis('census_tract').reset_index(name='buildings_count')
```

```python
buildings_by_mtracts.head()
```

```python
buildings_by_mtracts[:20].plot.bar(figsize=(20,4),
                             x='census_tract',
                             y='buildings_count')
```

```python
buildings_by_mtracts.info()
```

Adding the numbers of buildings back to the censustracts dataset for each borough.

```python
manhattantracts = manhattantracts.merge(buildings_by_mtracts, on='census_tract')
brooklyntracts = brooklyntracts.merge(buildings_by_brtracts, on='census_tract')
bronxtracts = bronxtracts.merge(buildings_by_bxtracts, on='census_tract')
statenislandtracts = statenislandtracts.merge(buildings_by_sitracts, on='census_tract')
queenstracts = queenstracts.merge(buildings_by_qtracts, on='census_tract')
```

```python
for borough in [manhattantracts, brooklyntracts, bronxtracts, statenislandtracts, queenstracts]:
    borough.plot(column='buildings_count', figsize=(10,10), scheme='quantiles', legend=True)
```
